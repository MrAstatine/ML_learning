```python
#house pricing dataset
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=fetch_california_housing()
```

```python
dataset=pd.DataFrame(df.data)
dataset.columns=df.feature_names
print(dataset.head())
```

```python
dataset['Price'] = df.target
dataset.head()
```

```python
#dividing the dataset into independent and dependent features
X=dataset.iloc[:,:-1] ##independent features
y=dataset.iloc[:,-1] ##dependent feature
'''
in the line above [:,:-1] means the first argument says that all records and the second argument says all columns except the last one, which is the target variable.
in the line above [:,-1] means the first argument says that all records and the second argument says the last column, which is the target variable.
'''
X.head()
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
linear_reg = LinearRegression()
mse= cross_val_score(linear_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
print(mse)
''' 
in the mse line
first u give the model, then the independent features, then the dependent feature, then the scoring method and finally the number of folds for cross-validation.
scoring='neg_mean_squared_error' means that the mean squared error will be returned as a negative value, so that higher values are better (since cross_val_score returns the negative of the score).
cross_val_score will return the negative mean squared error for each fold, so you can take the mean of these values to get the overall performance of the model.    
cv=5 means that the dataset will be divided into 5 folds for cross-validation. cross-validation is a technique used to evaluate the performance of a model by splitting the dataset into multiple subsets (folds) and training the model on different combinations of these subsets.
we could have done train_test_split and then fit the model on the training set and evaluate it on the test set, but cross-validation is a better way to evaluate the performance of a model as it uses all the data for training and testing.
'''
linear_reg.fit(X_train, y_train)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
#ridge regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params={'alpha':[1e-15, 1e-10, 1e-8,1e-3,1e-2,1,5,10,20,30,40,50,100,200,500,1000]}
#alpha is the regularization parameter for ridge regression, which controls the amount of regularization applied to the model. A higher value of alpha means more regularization, which can help prevent overfitting.
#GridSearchCV is used to find the best hyperparameters for the model by trying out different combinations of hyperparameters and evaluating the model's performance using cross-validation.

ridge_reg=GridSearchCV(ridge, params,scoring='neg_mean_squared_error', cv=5)
ridge_reg.fit(X_train, y_train)
print(ridge_reg.best_params_)
print(ridge_reg.best_score_)
#the best_params_ attribute will return the best hyperparameters found by GridSearchCV, and the best_score_ attribute will return the best score achieved by the model with those hyperparameters.
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso= Lasso()
params={'alpha':[1e-15, 1e-10, 1e-8,1e-3,1e-2,1,5,10,20,30,40,50,100,200,500,1000]}
lasso_reg=GridSearchCV(lasso, params,scoring='neg_mean_squared_error', cv=5)
lasso_reg.fit(X_train, y_train)
print(lasso_reg.best_params_)
print(lasso_reg.best_score_)
```

```python
from sklearn.metrics import  r2_score

y_pred_lin=linear_reg.predict(X_test)
r2_score_lin=r2_score(y_pred_lin, y_test)
print(r2_score_lin) 

y_pred_rid=ridge_reg.predict(X_test)
r2_score_rid=r2_score(y_pred_rid, y_test)
print(r2_score_rid) 

y_pred_las=lasso_reg.predict(X_test)
r2_score_las=r2_score(y_pred_las, y_test)
print(r2_score_las) 
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
df=load_breast_cancer()
X=pd.DataFrame(df['data'], columns=df['feature_names'])
#X.head()
y=pd.DataFrame(df['target'], columns=['Target'])
#y.head()
y['Target'].value_counts()
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
params=[{'C':[1,5,10]},{'max_iter':[100,200,300]}]
model1=LogisticRegression(C=100,max_iter=100)
model=GridSearchCV(model1, param_grid=params, scoring='f1', cv=5)
model.fit(X_train, y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
